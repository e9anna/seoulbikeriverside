```python
import folium as g
```

```python
import pandas as pd
df = pd.read_excel('2019top.xlsx')
#2019 주말/평일 이용객 수 Top10 정류장 
```

```python
df.head()
```

```python
lat = df['위도'].mean()
long = df['경도'].mean()
```

```python
m = g.Map([lat,long],zoom_start=12)

for i in df.index:
    sub_lat =  df.loc[i,'위도']
    sub_long = df.loc[i,'경도']
    
    title = df.loc[i,'평일주말']
  
    #주말과 평일 둘다 속하면 빨간색으로 표시, default(주말 혹은 평일)는 녹색
    
    color = 'green'
    if df.loc[i,'평일주말'] == '주말' and '평일':
        color = "red"
        
    #지도에 동그라미로 데이터 찍기    
    g.CircleMarker([sub_lat,sub_long],color=color,radius = 5, tooltip=title).add_to(m)

#한글이 안나오는 오류로 html로 trouble shooting
m.save('https://github.com/e9anna/seoulbikeriverside.git')
m# 자료를 바로 볼려면 : http://nbviewer.org/github/e9anna/seoulbikeriverside/blob/e03ba55c5f2118be1124a086dbb7f637db3c2324/map_simple_1028.ipynb
```
